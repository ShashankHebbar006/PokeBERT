```python
def seperate_name_type(input_list:list) -> tuple:
    # print(input_list)
    name_list = []
    type_list = []

    for i in range(len(input_list)):
        value = input_list[i]
        # print(value)
        pattern = "(Fire|Water|Grass|Electric|Ice|Dragon|Flying|Rock|Ground|Steel|Bug|Psychic|Dark|Ghost|Fairy|Fighting|Poison|Normal)"
        match = re.search(pattern,value)
        if match:
            # print(" ---> ", match.group())
            temp = match.group()
            name = value.replace(temp,"")
            if name:
                print(i," ---> ",name, temp)
        
            print("NAME ---> ", name)
            print("TYPE ---> ", temp)

            if name:
                name_list.append(name)
            type_list.append(temp)
        else:
            name_list.append(value)

    name = " ".join(name_list)
    type_str ="/".join(type_list)
    return (name, type_str)
```

```python
stat_dict = {"ID":[],
             "Name":[],
             "Type":[],
             "HP":[],
             "Attack":[],
             "Defense":[],
             "Sp. Attack":[],
             "Sp. Defense":[],
             "Speed":[]}

for i in range(df.shape[0]):
    data = df.loc[i]["#"]
    # print(data)
    data_split = data.split()
    # print(data_split[1:-7])
    stats_list = data.split()[-6:]
    
    id = data_split[0]
    # input_list = seperate_name_type(data_split[1:-7])
    name, type = seperate_name_type(data_split[1:-7])
    hp = stats_list[0]
    atk = stats_list[1]
    dfs = stats_list[2]
    sp_atk = stats_list[3]
    sp_def = stats_list[4]
    spd = stats_list[5]

    stat_dict["ID"].append(id)
    stat_dict["Name"].append(name)
    stat_dict["Type"].append(type)
    stat_dict["HP"].append(hp)
    stat_dict["Attack"].append(atk)
    stat_dict["Defense"].append(dfs)
    stat_dict["Sp. Attack"].append(sp_atk)
    stat_dict["Sp. Defense"].append(sp_def)
    stat_dict["Speed"].append(spd)

    # print("# ---> ", id)
    # print("Name ---> ", name)
    # print("Type ---> ", type)
    # # print("Stats ---> ", stats_list)
    # print("HP ---> ", hp)
    # print("Attack ---> ", atk)
    # print("Defense ---> ", dfs)
    # print("Sp. Attack  ---> ", sp_atk)
    # print("Sp. Defense ---> ", sp_def)
    # print("Speed ---> ", spd)
    print("-----------------------------------------------------------")
```

```python
poke_data = pd.DataFrame(stat_dict)
```
